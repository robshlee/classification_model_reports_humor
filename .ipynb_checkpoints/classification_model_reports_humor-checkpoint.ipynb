# Classification Models on Humor Styles Data

In this notebook I'll be building a binary classification model and build reports to interpret model performance. The README in the repo gives an explanation to this humor styles dataset.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

%matplotlib inline
```

```python
df = pd.read_csv('.//hsq_data.csv')
print(df.shape)
df.head()
```

## EDA

```python
# Check number of missings
for col in df[df.columns[0:32]]:
    df[col] = df[col].map(lambda x: np.nan if x==-1 else x)

df.isnull().sum()
```

```python
# Drop missings
df.dropna(inplace=True)
print('current shape: ', df.shape)
```

```python
# Check unique ages and gender counts
print('Unique ages:')
print(df['age'].unique())
print('')
print('Gender counts:')
print(df['gender'].value_counts())
```

```python
# Drop rows with age >= 100
df = df[df['age'] < 100]

# Drop rows where gender is not male or female and change to binary
# 0 = male, 1 = female
df = df[df['gender'].isin([1,2])]
df['gender'].map(lambda x: 0 if x==1 else 0 )

print('current shape: ', df.shape)
```

We'll make the aggressive score our target.

```python
# Aggressive is spelled incorrectly, so we'll fix that
df.rename(columns={'agressive':'aggressive'}, inplace=True)
```

```python
# Set up feature matrix and target vector
X = df.drop('aggressive', axis=1)
y = df['aggressive']

X_train, X_test, y_train, y_test = train_test_split(X, y, \
                                test_size=0.2, random_state=1337)
```

```python
# See summary statistics for aggressive column
y_train.describe().to_frame()
```

```python
# See distribution of aggressive column
print(y_train.value_counts().to_frame().sort_index())

plt.hist(y_train)
```

Let's make this a binary classification problem. 0 for aggressive scores <= 3 and 1 for aggressive scores > 3.

```python
y_train = y_train.map(lambda x: 1 if x > 3 else 0)
y_test = y_test.map(lambda x: 1 if x > 3 else 0)

print('Count of aggressive > 3: ', y_train.value_counts()[1])
print('Count of total: ', len(y_train))
print('% of aggressive > 3: ', y_train.mean())
```

Baseline accuracy is 36.75%.

```python
pd.DataFrame(X_train, y_train)
```

## Modeling

```python
logreg = LogisticRegression()

cross_val_score(logreg, X_train, y_train, cv=5).mean()
```

Base logistic regression has a cross-validated accuracy of 94.45% which is really good. It beat out baseline by a large margin.

```python
logreg.fit(X_train, y_train)
logreg.score(X_test, y_test)
```

The model performed even better on the test data. 

```python
# Classification report and confusion matrix
yhat = logreg.predict(X_test)

print(classification_report(y_test, yhat))

pd.DataFrame(confusion_matrix(y_test, yhat), columns=['predicted > 3',\
            'predicted <= 3']).rename(index={0: 'is > 3', 1: 'is <= 3'})
```

Our model has slightly higher precision on aggresive scores <= 3 than it did predicting > 30. That being said it was by a very small margin and with 97.45% accuracy on the test data, I feel good about the results. With our model predicting scores <= 3 incorrectly 4 times and predicting scores > 3 incorrectly once I won't waste time trying to balance the FPR and FNR.

```python
# Plot ROC curve
yhat_pp = logreg.predict_proba(X_test)
fpr, tpr, _ = roc_curve(y_test, yhat_pp[:,1])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=[8,8])
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc, linewidth=4)
plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([-0.05, 1.0])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.title('ROC: Aggresive score > 3', fontsize=18)
plt.legend(loc="lower right")
plt.show()
```
